```python
def exp_hamiltonians(seqs, J, h):
    if len(seqs.shape) ==3:
        seqs = seqs.reshape(seqs.shape[0], -1) # in case the onehots arent already flat. 
    J = np.moveaxis(J, 1,2)
    J = J.reshape(seqs.shape[-1], seqs.shape[-1])
    h = h.reshape(-1)
    #print(J.shape, h.shape, seqs.shape)
    H = ((seqs.T*(J@seqs.T))/2 ).T.sum(1) + seqs@h
    return H # before it was an array of an array.
```

```python
def hard_max(inp):
    if inp.shape[1] > L: # if it isnt a sequence of ints, then all to one hot.  
        #print('the input to oh is', inp)

        if len(inp)>2:
            inp = inp.reshape(inp.shape[0], -1, 20)
        else:
            inp = inp.reshape(-1,20)

        inp = np.argmax(inp, axis=-1)
    return inp
```

```python
def n_to_aa(inp):
    to_aa = []
    N=20 # none of these focus have gaps, else should be 21. 
    for seq in inp:
        st = ''
        for s in seq:
            st+= n2a[s]
        to_aa.append(st)
    to_aa=np.asarray(to_aa)
    return to_aa
```
